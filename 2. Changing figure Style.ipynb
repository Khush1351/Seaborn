```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 
  
# loading dataset 
data = sns.load_dataset("iris")

# changing the theme to darkgrid
sns.set_style(style="darkgrid")

# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 

plt.show()
```
